---

create new dataframe with all the stat calculations every 3 days

```python
from scipy.stats import linregress
import math

path_final = '/content/drive/My Drive/SUMMER 22/intern/analyzed_data.csv'

#index, start day, end day, slope

x_vals = new_clean_dataset['days'].tolist()
y_vals = new_clean_dataset['water usage per day (mm)'].tolist()
dates = new_clean_dataset['time stamp'].tolist()

new_head = ['Start Day', 'End Day', 'Slope (3 day)', 'Slope (all points)', 'Stage of Growth', 'Days in Growth Stage']
#days in growth stage starts with start date
start_days = []
end_days = []
slope_3 = []
slope_all = []
stages = []
days_in_growth = []
days = 0

start_index = 0 #different for every stage
index = start_index
batch_interval = 2  #how many away from start point
dataframe_length = len(new_clean_dataset)
stats_df = pd.DataFrame(columns = new_head)

while(index + batch_interval < dataframe_length):
  index_2 = index + batch_interval

  temp_x = [x_vals[index], x_vals[index_2]]
  temp_y = [y_vals[index], y_vals[index_2]]
  slope, intercept, r_value, p_value, stderr = linregress(temp_x, temp_y) #slope of regression line (3 day)

  final_x = [x_vals[start_index], x_vals[index_2]]
  final_y = [y_vals[start_index], y_vals[index_2]]

  f_slope, f_intercept, f_r_value, f_p_value, f_stderr = linregress(final_x, final_y) #slope of regression line

  days += 1
  days_in_growth.append(days)

  #what stage
  if (abs(0 - slope) <= 0.1 and y_vals[index] < 5):
    stage = 'Germination Stage'

  if (slope > 0.1 and slope < 1):
    stage = 'Vegitative Growth Stage'

    if(abs(0 - f_slope) <= 0.1 and stages[index - 1] == 'Germination Stage'):
      stage = 'Germination Stage - Vegitative Growth Stage'
      start_index = index+1
      days = 0

  if(abs(0 - slope) <= 0.1 and y_vals[index] > 10):
    stage = 'Reproductive Stage'

    if(abs(0 - f_slope) <= 1 and stages[index - 1] == 'Vegitative Growth Stage'):
      stage = 'Vegitative Growth Stage - Reproductive Stage'
      start_index = index+1
      days = 0

  if(slope > -1 and slope < -0.1):
    stage = 'Maturity'

    if(f_slope < 0 and stages[index - 1] == 'Reproductive Stage'):
      stage = 'Reproductive Stage - Maturity'
      start_index = index+1
      days = 0

  stages.append(stage)


  #CODE CONFIDENCE
  
  start_days.append(dates[index])
  end_days.append(dates[index_2])
  slope_3.append(slope)
  slope_all.append(f_slope)



  index += 1

stats_df = stats_df.append(pd.DataFrame(list(zip(start_days, end_days, slope_3, slope_all, stages, days_in_growth)), columns = new_head))
stats_df = stats_df.set_index(pd.Index(list(range(1, len(stats_df)+1))))

  
stats_df.to_csv(path_final)

stats_df
```
